```python
def frame_image(img, frame_width):
    b = frame_width # border size in pixel
    ny, nx = img.shape[0], img.shape[1] # resolution / number of pixels in x and y
    if img.ndim == 3: # rgb or rgba array
        framed_img = np.ones((b+ny+b, b+nx+b, img.shape[2]))
    elif img.ndim == 2: # grayscale image
        framed_img = np.ones((b+ny+b, b+nx+b))
    framed_img[b:-b, b:-b] = img
    return framed_img

def format_pop(pop):
    if pop > 1e6:
        div = 1e6
        s = 'M'
    else:
        div = 1e3
        s = 'k'
    return "%2.1f%s"%(float(pop)/div, s)
```

```python
from scipy.stats import wasserstein_distance

def compute_effort_emd(source, target):
    return wasserstein_distance(source, target)

def compute_discrepancy(source, target):
    dif = source - target
    tot_disc = dif.sum()/float(target.sum())
    return tot_disc

colormap = {"bldg":"black", "pop":"green", "lum":"orange", 
            "all":"blue", "bnds":"gray"}

def get_profiles(city_info):
    prof = {}
    for s in ['bldg', 'pop', 'lum', 'bnds']:
        mu = city_info['profiles'][s][0][2:]
        scale = mu.max() #v['sum area'][s]
        p = mu / scale
        p[np.isnan(p)] = 0
        prof[s] = p
    return prof

def plot_profiles(city_info, a, title=None):
    prof = get_profiles(city_info)
    for s in ['bldg', 'pop', 'lum', 'bnds']:
        a.plot(distance_km[2:], prof[s], lw=3, label=s, color=colormap[s])
     # a.legend(loc="best")
    if title is not None:
        a.set_title(title)
        
def compute_stats_profiles(city_info):
    prof = get_profiles(city_info)
    dis_bld = compute_discrepancy(prof['bldg'], prof['pop'])
    dis_lum = compute_discrepancy(prof['lum'], prof['pop'])
    emd_bld = compute_effort_emd(prof['bldg'], prof['pop'])
    emd_lum = compute_effort_emd(prof['lum'], prof['pop'])
    return dis_bld, dis_lum, emd_bld, emd_lum
    
    
from scipy.stats import linregress

def f_ols(y):
    g, c, R, p, s = linregress(distance_km,y)
    return g, c, R**2, p, s
```

```python
from sklearn.cluster import KMeans, MiniBatchKMeans
from numpy.random import random_sample
from scipy.spatial.distance import cdist, pdist

def kmeans_var_expl(data, n):
    kMeansVar = [MiniBatchKMeans(n_clusters=k, batch_size=1000).fit(data) for k in range(1, n)]
    centroids = [X.cluster_centers_ for X in kMeansVar]
    k_euclid = [cdist(data, cent) for cent in centroids]
    dist = [np.min(ke, axis=1) for ke in k_euclid]
    wcss = [sum(d**2) for d in dist]
    tss = sum(pdist(data)**2)/data.shape[0]
    bss = tss - wcss
    return tss, bss, wcss
```

```python
def plot_clusters(X, C, centroids, 
                  nrow=1, text="", figsize=(6,4), top=0.75,
                  xlabels=True):
    K = len(np.unique(C))
    clust_sizes = pd.Series(C).value_counts().values
    nx = nrow
    ny = K / nx
    fig, ax = plt.subplots(nx,ny, 
                           gridspec_kw={"wspace":0.15, "hspace":0.3},
                           figsize=figsize, sharex=True, sharey=True)
    if len(ax.shape)==1:
        ax = np.expand_dims(ax, axis=0)
    for k,a in zip(range(K), ax.flatten()):
        idx = np.where(C == k)[0]
        std = X[idx,:].std(0)
        a.fill_between(distance_km,
           centroids[k,:]-std, 
           centroids[k,:]+std, 
           color=colormap[text], alpha=0.3)
        a.plot(distance_km, centroids[k,:], lw=5, color=colormap[text])
        # a.axis("off")
        a.set_title("C%d: %2.1f%%"%(k,100*float(clust_sizes[k])/clust_sizes.sum()))
        if k / ny == nx-1:
            if xlabels: a.set_xlabel("d [km]")
            a.set_xticks(distance_km[::6])
            a.set_xticklabels(distance_km[::6], rotation=30)
        if k % ny == 0:
            a.set_ylabel("$x_{\cdot}(d)$")

    plt.suptitle("Classes of normalized profiles $x_{\cdot}(d)$: centroids")
    fig.subplots_adjust(top=top)
```

#### Generate interpolation map

```python
def generate_interpolation_map(Z,M):
    N, nz = Z.shape
    Z_map = np.zeros((N-1,M+1,nz))
    for n in range(N-1):
        dZ = (Z[n+1,:] - Z[n,:]) / float(M)
        for m in range(M+1):
            Z_map[n,m,:] = Z[n,:] + m*dZ
        Z_map[n,0,:] = Z[n,:]
        Z_map[n,m,:] = Z[n+1,:]
    Z_map = Z_map.reshape(((N-1)*(M+1), nz))
    return Z_map
```

```python
from skimage.transform import resize

orig_data_path = "/home/data/world-cities/spatial-maps/samples"

z_rec_dict = {}
for i,f in enumerate(filenames):
    city_name = cities_sel.values()[i]['city']
    country_name = cities_sel.values()[i]['country']
    sample_name = cities_sel.keys()[i]
    city_info_real = cities_sel[sample_name]
    clear_output(wait=True)
    print sample_name

    # read in real image
    img_real_orig = imread("%s/%s.tif"%(orig_data_path, sample_name))
    img_real = imread(f)
    img_real = basic_preprocess(img_real, width, log=take_log, normalize=normalize)[...,:3]

    # recover latent vector
    z_rec, loss_rec = reverse_z(netG.cuda(), np.expand_dims(img_real.transpose((2,1,0)),0),
                                    nz=z_num, cuda=True, clip='stochastic', 
                                    lr=0.0002, niter=30000, loss_type='L2',
                                    apply_transform=norm_transf)
    img_rec = netG(z_rec).data.cpu().numpy().squeeze().transpose((1,2,0))
    img_rec = basic_preprocess(img_rec, width, normalize=True)
    z_rec_dict[sample_name] = z_rec
    
    # save image and latent space data
    sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 1})
    fig, ax = plt.subplots(2,5, figsize=(14,6), 
                           gridspec_kw={"wspace":0.05, "hspace":0.05})
    ax[0,0].imshow(np.flip(np.rot90(img_real[...,:3],k=3),1), cmap=cm.GnBu)
    ax[1,0].imshow(np.flip(img_rec[...,:3],0), cmap=cm.GnBu)
    for c,img in enumerate([img_real, img_rec]):
        ax[c,0].set_ylabel(city, rotation=90); 
        ax[c,0].set_xticks([]); ax[c,0].set_yticks([]); 
    for c in range(3):
        ax[0,c+1].imshow(np.flip(np.rot90(img_real[...,c],k=3),1), cmap=cm.GnBu); 
        ax[1,c+1].imshow(np.flip(img_rec[...,c],0), cmap=cm.GnBu); 
        ax[0,c+1].set_title(channel_names[c])
        _ = ax[0,c+1].axis("off"); 
        _ = ax[1,c+1].axis("off"); 

    ax[0,0].set_ylabel("real")
    ax[1,0].set_ylabel("recovered")
    fig.subplots_adjust(top=0.82)
    fig.suptitle("%s, %s"%(city_name, country_name))

    # compute radial profile for recovered image
    water = img_real_orig[...,3]
    bnds = img_real_orig[...,4]
    img_rec_proc = np.flip(resize(img_rec, (128,128)),0)
    city_info_rec = compute_profiles(img_rec_proc, water, bnds, sample_name)
    
    # plot profiles
    plot_profiles(city_info_real, ax[0,-1], title="$\hat x_{\cdot}(d)$"); _ = ax[0,-1].axis("off"); 
    plot_profiles(city_info_rec, ax[1,-1]); _ = ax[1,-1].axis("off"); 
    
    # compute discrepancy for bld, lum for real and recovered images
    mu_real = get_profiles(city_info_real)
    mu_rec  = get_profiles(city_info_rec)
    
    discr_real = {k:compute_discrepancy(mu_real[k], mu_real['pop']) \
                   for k in ['bldg', 'lum']}
    discr_rec  = {k:compute_discrepancy(mu_rec[k], mu_rec['pop']) \
                   for k in ['bldg', 'lum']}
    
    fig.savefig(experiment_dir + "/recovered/%s_combined.png"%sample_name, dpi=500)
    # plt.close(fig) 
    imsave(experiment_dir + "/recovered/%s_real.png"%sample_name, img_real)
    imsave(experiment_dir + "/recovered/%s_recovered.png"%sample_name, img_rec)
    np.savetxt(experiment_dir + "/recovered/%s_latent.txt"%sample_name, 
                  z_rec.data.cpu().numpy().squeeze())
    
```
